```python
def most_common(col):
    most_cmn=pd.get_dummies(col).sum().sort_values(ascending = False).index[0]
    return most_cmn
```

```python
def replace_missing(col):
    most_cmn = most_common(col)
    for i in range(len(col)):
        if pd.isnull(col[i]):
            col[i] = most_cmn
            inplace=True
        
```

```python
def feature_comp_plot(col1,col2):
    
    fig, ax = plt.subplots(figsize=(6,4)) 
    
    x2 = data.groupby([col1,col2])['Loan_ID'].count()
    x2 = x2.reset_index()
    
    stats = x2[col2].drop_duplicates()
    margin_bottom = np.zeros(len(x2[col1].drop_duplicates()))
    colors = ["#006D2C", "#31A354","#74C476"]
    
    for num, status in enumerate(stats):
        values = list(x2[x2[col2] == status].loc[:, 'Loan_ID'])
        x2[x2[col2] == status].plot.bar(x=col1,y='Loan_ID', ax=ax, stacked=True, 
                                    bottom = margin_bottom, color=colors[num], label=status)
        margin_bottom += values
    ax.set_title(col1+" vs. "+col2)
    plt.savefig(col1+" vs. "+col2+".jpeg",bbox_inches = 'tight')
    plt.show()
    
```

### Applying Inferential Statistics

```python
def bootstrap_replicate_1d(data, func):
    bs_sample = np.random.choice(data, len(data))
    return func(bs_sample)

def draw_bs_reps(data,func,size=1):
    bs_replicates = np.empty(size)
    for i in range(size):
        bs_replicates[i] = bootstrap_replicate_1d(data,func)
    return bs_replicates
    
```

```python
np.random.seed(47)
N_rep = 10000

bs_replicates  = draw_bs_reps(data.LoanAmount,np.mean,N_rep)
conf_int = np.percentile(bs_replicates,[5])
conf_int
```

```python
# loan not approved
loan_n = data.loc[data.Loan_Status == 'N']

# loan approved
loan_y = data.loc[data.Loan_Status == 'Y']
```

```python
np.random.seed(47)
N_rep = 10000

nloan_replicates = draw_bs_reps(loan_n.LoanAmount,np.std,N_rep)
yloan_replicates = draw_bs_reps(loan_y.LoanAmount,np.std,N_rep)

diff_of_reps = yloan_replicates - nloan_replicates

conf_int = np.percentile(diff_of_reps,[2.5,97.5])
```

```python
lower, upper = conf_int
print("The 95% confidence interval is between", round(lower,2), "and", round(upper,2))
```

```python
axvline_kwargs = {"linestyle" : "dashed", "linewidth" : "1"} 
diff_of_reps = pd.Series(diff_of_reps)
diff_of_reps.hist(bins=20, grid=False,
           figsize=(14,8), xlabelsize=10, edgecolor='k')
plt.xlabel("Dollars", fontsize=15)
plt.ylabel("Frequency",fontsize=15)
plt.axvline(np.percentile(diff_of_reps, [97.5]), **axvline_kwargs, label="2.5%")
plt.axvline(diff_of_reps.mean(), **axvline_kwargs, color="red", label="mean")
plt.axvline(np.percentile(diff_of_reps, [2.5]), **axvline_kwargs, label="97.5%")
plt.legend(loc='upper right')
plt.show()
```

#### Confidence interval and p-value

```python
# loan not approved
loan_n = data.loc[data.Loan_Status == 'N']
nloan_amnt = loan_n.LoanAmount.values

# loan approved
loan_y = data.loc[data.Loan_Status == 'Y']
yloan_amnt = loan_y.LoanAmount.values
```

```python
emp_diff = np.mean(yloan_amnt) - np.mean(nloan_amnt)

mean_amnt = np.mean(data.LoanAmount)

nloan_shifted = nloan_amnt - np.mean(nloan_amnt) + mean_amnt
yloan_shifted = yloan_amnt - np.mean(yloan_amnt) + mean_amnt
```

```python
np.random.seed(47)
N_rep =10000

nloan_shifted_reps = draw_bs_reps(nloan_shifted,np.mean,N_rep)
yloan_shifted_reps = draw_bs_reps(yloan_shifted,np.mean,N_rep)

diff_reps = yloan_shifted_reps - nloan_shifted_reps

p_val = np.sum(diff_reps >= emp_diff)/len(diff_reps)
print('p-value =', p_val)
```

```python
axvline_kwargs = {"linestyle" : "dashed", "linewidth" : "1"} 

plt.figure(figsize=(12,8))

plt.hist(nloan_shifted_reps, bins=20, label="no-loan means")
plt.hist(yloan_shifted_reps, bins=20, label="loan means")
plt.hist(diff_reps, bins=20, label="difference in means")
plt.xlabel("Dollars", fontsize=15)
plt.ylabel("Frequency",fontsize=15)
plt.axvline(emp_diff, **axvline_kwargs, label="empirical mean difference")
plt.axvline(mean_amnt, **axvline_kwargs, color="k", label="mean of all loan amount")
plt.legend(loc='best')
plt.show()
```

```python
#Data Preprocessing
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_scaled = pd.DataFrame(sc_X.fit_transform(X), columns=X.columns)
```

```python
#split train and test sets

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, random_state=0)
```

```python
#import classifier
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_train,y_train)
y_pred = classifier.predict(X_test)
classifier.score(X_test,y_test) #classifier performance on test set
```

```python
# importing performance measuring tools
from sklearn.metrics import accuracy_score, confusion_matrix,recall_score,precision_score,classification_report

recall_score(y_test,y_pred,average='macro')
```

```python
cr=classification_report(y_test,y_pred)
print(cr)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
precision_score(y_test,y_pred,average='macro')
```

```python
#import KNeighborsClassifier
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=21,weights='distance',p=1)
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()
model.fit(X_train,y_train)
model.score(X_test,y_test)
param_dict=({'n_neighbors':range(3,11,2),'weights':['uniform','distance'],'p':[1,2,3,4,5]})
from sklearn.model_selection import GridSearchCV
best_model=GridSearchCV(model,param_dict,cv=5)
best_model.fit(X_scaled,y)
best_model.best_params_
best_model.best_score_
```

```python
from sklearn.ensemble import RandomForestClassifier
model2 = RandomForestClassifier(max_depth=25)
model2.fit(X_train,y_train)
model2.score(X_test,y_test)
param_dict_2=({'n_estimators':range(2,50)})
from sklearn.model_selection import GridSearchCV
best_model=GridSearchCV(model2,param_dict_2,cv=5)
best_model.fit(X_scaled,y)
best_model.best_params_
best_model.best_score_
```

```python
from sklearn.ensemble import AdaBoostClassifier
model3 = AdaBoostClassifier(n_estimators=20)
model3.fit(X_train,y_train)
model3.score(X_test,y_test)
param_dict_3=({'n_estimators':range(2,50)})
from sklearn.model_selection import GridSearchCV
best_model=GridSearchCV(model3,param_dict_3,cv=5)
best_model.fit(X_scaled,y)
best_model.best_params_
best_model.best_score_
```

```python
#Support vector Machine model
from sklearn.svm import SVC
model_svc = SVC(kernel='linear',gamma=0.001,C=1.0)
model_svc.fit(X_train,y_train)
model_svc.score(X_test,y_test)
```

```python
#Estimating the best model using Cross-validation
new_model=best_model.best_estimator_ #gives the best model estimation
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
cross_val_score(new_model, X_scaled,y,cv=5).mean()
```
